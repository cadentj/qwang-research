```python
import osmnx as ox
import pandas as pd
```

```python
PLACE_NAME = "Houston, Texas, United States"
houston_geo = ox.geometries_from_address(PLACE_NAME, tags={'amenity':True, 
                                                      'office':True,
                                                      'shop':True,
                                                      'leisure':True,
                                                      })

# Some areas have no office buildings
```

```python
houston_geo = houston_geo.loc[houston_geo['name'] != 'Buffalo Bayou Park']

houston_amenity = houston_geo[houston_geo['amenity'].notnull()]
houston_shop = houston_geo[houston_geo['shop'].notnull()]
houston_leisure = houston_geo[houston_geo['leisure'].notnull()]
try:
    houston_office = houston_geo[houston_geo['office'].notnull()]
except: 
    print("There are no offices in this area.")
```
