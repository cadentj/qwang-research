```python
# TODO: FINISH LMFAO

def eigenvector_centrality(graph) :
    graph_adj = nx.adjacency_matrix(graph).toarray()
    # Skipping the steps of actually calculating eigenvalues and vectors and using inbuilt methods instead

    e_val, e_vec = np.linalg.eig(graph_adj)
    max = np.argmax(e_val)
    centrality = e_val[max] * e_vec[max] 
    return centrality / np.linalg.norm(centrality)
    
centrality = eigenvector_centrality(graph)
for a in centrality: print("{:.3e}".format(a))

nx.eigenvector_centrality(graph)
```
